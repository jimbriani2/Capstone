```python
import pandas as pd
import sqlite3
```

```python
Pitchers_Data = pd.read_csv("../CapstoneProject/PitcherData.csv")
```

```python
conn = sqlite3.connect("Pitcher_Data.db")
```

```python
Pitchers_Data.to_sql('Pitchers', conn)
```

```python
sql_query = 'SELECT * FROM Pitchers'
Pitchers_Data = pd.read_sql(sql_query, conn)
print(Pitchers_Data)
```

```python
sql_query = 'SELECT * FROM Pitchers WHERE Starts > 25 AND Year > 2015 AND ERA < 3.5'
Pitchers_Data = pd.read_sql(sql_query, conn)
print(Pitchers_Data)
```

```python
sql_query = 'SELECT * FROM Pitchers WHERE Games > 40 AND Year > 2015 AND ERA < 4'
Pitchers_Data = pd.read_sql(sql_query, conn)
print(Pitchers_Data)
```
